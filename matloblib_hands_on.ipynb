### Pandas Styles ###

```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['prce'] = boston['target']
```

```python
hdf = df.head(10)
hdf
```

```python
hdf.style.highlight_max()
```

```python
hdf.style.background_gradient()
```

```python
hdf.style.bar()
```

```python
def odd_green(val):
    if val % 2 !=0:
        return 'color:green'
    return ''
```

```python
hdf.style.applymap(odd_green)
```

```python
def orange10(val):
    if val>10:
        return 'background-color: orange'
    return ''
```

```python
hdf.style.applymap(odd_green).applymap(orange10)
```

```python
hdf.style.applymap(odd_green, subset=['ZN', 'RAD'])
```

```python
def top20(col):
    is_top20 = col > col.quantile(.8)
    return ['font-weight: bold' if v else '' for v in is_top20]
```

```python
hdf.style.apply(top20)
```

### MPL ###

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
plt.plot(xs, ys)
```

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.legend()
```

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs),label='sin(x)')
plt.legend()
```

```python
plt.subplot(2, 1, 1)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()
plt.subplot(2, 1, 2)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
```
